### Tolkning av h(y)
$$
\frac{1}{\sqrt{4 \pi Dt}} e ^{-\frac{(x-y)^2}{4Dt}} \sim \mathcal{N} (y, 2Dt)
$$
$$
t \longrightarrow 0 \Longrightarrow \mathcal{N} (y, 2Dt) \longrightarrow \delta (x-y)

# Oppgave 2a)
I koden brukes bV for å tilsvare $\beta V(x)$, bV_pos = $\beta V(x+1)$, bV_neg = $\beta V(x-1)$

```python
from collections.abc import Sequence


import numpy as np
import matplotlib.pyplot as plt
from numba import njit


def plot_distributions(
    distributions: Sequence[np.ndarray],
    labels: Sequence[str],
    title: str,
    initital_distribution: np.ndarray,
):
    """Hjelpefunksjon for å plotte flere distribusjoner i tillegg til en initiell distribusjon."""
    plt.figure()
    plt.hist(
        initital_distribution,
        bins=100,
        density=True,
        label="Initiell distribusjon av partikler",
        color="#22222299",
    )
    for dist, label in list(zip(distributions, labels)):
        plt.hist(dist, bins=50, density=True, label=label, alpha=0.5)
    plt.legend()
    plt.title(title)
    plt.xlabel("Position")
    plt.ylabel("Probability density")
    plt.show()


@njit
def potential_func_1(x: np.ndarray, beta_k: float = 1):
    return np.full_like(x, beta_k)


@njit
def potential_func_2(x: np.ndarray, beta_k: float = 1):
    return -beta_k * x


@njit
def prob_left(bV: np.ndarray, bV_pos: np.ndarray, bV_neg: np.ndarray):
    return 1 / (1 + np.exp(-(bV_pos - bV_neg)) + np.exp(-(bV - bV_neg)))


@njit
def prob_right(bV: np.ndarray, bV_pos: np.ndarray, bV_neg: np.ndarray):
    return 1 / (1 + np.exp(-(bV_neg - bV_pos)) + np.exp(-(bV - bV_pos)))


N = 200


@njit
def walk_1(x: np.ndarray, beta_k: float = 1) -> np.ndarray:
    """Virrevandrer partiklene i x med potensialet konstant = beta*k"""
    _walk_dist_random = np.random.uniform(0, 1, size=(N, x.shape[0]))
    for walk_dist_random in _walk_dist_random:
        bV = potential_func_1(x, beta_k)
        bV_pos = potential_func_1(x + 1, beta_k)
        bV_neg = potential_func_1(x - 1, beta_k)
        x -= np.where(walk_dist_random < prob_left(bV, bV_pos, bV_neg), 1, 0)
        x += np.where(walk_dist_random > (1 - prob_right(bV, bV_pos, bV_neg)), 1, 0)
    return x


@njit
def walk_2(x: np.ndarray, beta_k: float = 1) -> np.ndarray:
    """Virrevandrer partiklene i x med potensialet = -beta*k*x"""
    _walk_dist_random = np.random.uniform(0, 1, size=(N, x.shape[0]))
    for walk_dist_random in _walk_dist_random:
        bV = potential_func_2(x, beta_k)
        bV_pos = potential_func_2(x + 1, beta_k)
        bV_neg = potential_func_2(x - 1, beta_k)
        x -= np.where(walk_dist_random < prob_left(bV, bV_pos, bV_neg), 1, 0)
        x += np.where(walk_dist_random > (1 - prob_right(bV, bV_pos, bV_neg)), 1, 0)
    return x


x_1 = np.array([])
x_2 = np.array([])
for _ in range(100):
    x_1 = np.concat([x_1, walk_1(np.arange(0, 100))])
    x_2 = np.concat([x_2, walk_2(np.arange(0, 100))])

plot_distributions(
    (x_1, x_2),
    ("Distribusjon av partikler med V(x)=k", "Distribusjon av partikler med V(x)=-kx"),
    "Virrevandring uten kollisjon",
    np.arange(0, 100),
)
```

Den grå boksen viser startposisjon til alle partiklene. Vi ser at partiklene som lever i et konstant potensial spres symmetrisk ut fra sin startposisjon, det kan se ut som sand som lager en haug. Derimot partiklene som opplever potensial lik $-\beta*k*x$ drifter som regel til høyre. Gjennomsnittlig går alle partiklene like mye til høyre, ettersom det ikke er noen mekanisme for å stoppe partiklene til venstre fra å gå til høyre. 

# Oppgave 2b)

```python
@njit
def walk_collision(
    x: np.ndarray, walk_left: np.ndarray, walk_right: np.ndarray, idxs: np.ndarray
):
    """Gå gjennom partiklene i tilfeldig rekkefølge og sjekk hvilken retning de prøver å gå og sjekk om de kan gå den retningen."""
    np.random.shuffle(idxs)
    for idx in idxs:
        if walk_left[idx] and not np.isin(x[idx] - 1, x):
            x[idx] -= 1

        if walk_right[idx] and not np.isin(x[idx] + 1, x):
            x[idx] += 1


@njit
def get_walk_arrs(
    bV: np.ndarray, bV_pos: np.ndarray, bV_neg: np.ndarray, walk_uniform: np.ndarray
):
    """Lag arrays som beskriver om en partikkel prøver å gå til enten høyre eller venstre."""
    _prob_left = prob_left(bV, bV_pos, bV_neg)
    _prob_right = prob_right(bV, bV_pos, bV_neg)
    walk_left = walk_uniform < _prob_left
    walk_right = walk_uniform > (1 - _prob_right)
    return walk_left, walk_right


@njit
def walk_collision_1(x: np.ndarray, beta_k: float = 1) -> np.ndarray:
    """Virrevandrer partiklene i x med potensialet konstant = beta*k, med kollisjoner"""
    _walk_dist_random = np.random.uniform(0, 1, size=(N, x.shape[0]))
    idxs = np.arange(x.shape[0])
    for walk_dist_random in _walk_dist_random:
        bV = potential_func_1(x, beta_k)
        bV_pos = potential_func_1(x + 1, beta_k)
        bV_neg = potential_func_1(x - 1, beta_k)
        walk_left, walk_right = get_walk_arrs(bV, bV_pos, bV_neg, walk_dist_random)
        walk_collision(x, walk_left, walk_right, idxs)
    return x


@njit
def walk_collision_2(x: np.ndarray, beta_k: float = 1) -> np.ndarray:
    """Virrevandrer partiklene i x med potensialet = -beta*k*x, med kollisjoner"""
    _walk_dist_random = np.random.uniform(0, 1, size=(N, x.shape[0]))
    idxs = np.arange(x.shape[0])
    for walk_dist_random in _walk_dist_random:
        bV = potential_func_2(x, beta_k)
        bV_pos = potential_func_2(x + 1, beta_k)
        bV_neg = potential_func_2(x - 1, beta_k)
        walk_left, walk_right = get_walk_arrs(bV, bV_pos, bV_neg, walk_dist_random)
        walk_collision(x, walk_left, walk_right, idxs)
    return x


x_1 = np.array([])
x_2 = np.array([])
for _ in range(100):
    x_1 = np.concat([x_1, walk_collision_1(np.arange(0, 100))])
    x_2 = np.concat([x_2, walk_collision_2(np.arange(0, 100))])


plot_distributions(
    (x_1, x_2),
    ("Distribusjon av partikler med V(x)=k", "Distribusjon av partikler med V(x)=-kx"),
    "Virrevandring med kollisjon",
    np.arange(0, 100),
)
```

Ser at fordelingen av partiklene i det konstante potensialet ser ganske lik ut som fordelingen i samme potensial uten kollisjon. Dette kan forklares ved at med kollisjoner kan partiklene ytterst kun gå videre utover, mens uten kollisjoner kan partiklene ytterst også gå innover i tillegg til at partikler som ikke er ytterst kan gå utover. Disse effektene kansellerer hverandre og vi får nesten samme fordeling. Når det gjelder partiklene som er i et potensial som avhenger av x ser vi tydelig at de fortsatt vil gå til høyre, men at partiklene som begynner til venstre ikke kan gå til høyre før alle partiklene til høyre har flyttet seg. Dermed vil partiklene som startet til høyre kunne gå lengst, mens partiklene som starter til venstre ikke kommer seg langt fra startposisjonen sin.

# Oppgave 3a)

```python
h = 1
N_x = 100
dt = 1
N_p = 12 * N_x


@njit
def potential_func_sawtooth(
    x: np.ndarray, beta_k: float = 1000, alpha: float = 0.8, N_x: float = 100
):
    """Potensialfunksjon for Ratchet potential."""
    x_pos = x % N_x
    return np.where(
        x_pos <= alpha * N_x,
        beta_k * x_pos / (alpha * N_x),
        -beta_k * (x_pos - N_x) / ((1 - alpha) * N_x),
    )


@njit
def wrap_around(x: np.ndarray, N_x: float = 100, size: int = 2):
    """Passer på at x holder seg til det definerte området."""
    x = np.where(x < 0, x + size * N_x, x)
    return np.where(x >= size * N_x, x - size * N_x, x)


def _walk_saw_tooth_current(
    x: np.ndarray,
    T_p: int = 500,
    cycles: int = 10,
    alpha: float = 0.8,
    beta_k: float = 1000,
    N_x: float = 100,
) -> np.ndarray:
    """Itererer gjennom tidssteg og oppdaterer posisjonen til partiklene ut i fra de utregnede sannsynlighetene."""
    _walk_dist_random = np.random.uniform(0, 1, size=(2 * T_p * cycles, x.shape[0]))
    current = np.zeros((2 * T_p * cycles))
    for step, walk_dist_random in enumerate(_walk_dist_random):
        t = step * dt
        if (t // T_p) % 2 == 0:
            diff = np.where((walk_dist_random < 1 / 3), -1, 0)
            diff += np.where((walk_dist_random > 2 / 3), 1, 0)
        else:
            bV = potential_func_sawtooth(x, beta_k, alpha, N_x)
            bV_pos = potential_func_sawtooth(x + 1, beta_k, alpha, N_x)
            bV_neg = potential_func_sawtooth(x - 1, beta_k, alpha, N_x)
            diff = np.where(walk_dist_random < prob_left(bV, bV_pos, bV_neg), -1, 0)
            diff += np.where(
                walk_dist_random > (1 - prob_right(bV, bV_pos, bV_neg)), 1, 0
            )
        current[step] = diff.mean()
        x += diff
        x = wrap_around(x)
    return current


def walk_saw_tooth_current(
    x: np.ndarray,
    T_p: int = 500,
    cycles: int = 10,
    alpha: float = 0.8,
    beta_k: float = 1000,
    N_x: int = 100,
) -> np.ndarray:
    """Regner ut cycle-averaged current for ratched potensial."""
    current = _walk_saw_tooth_current(x, T_p, cycles, alpha, beta_k, N_x)
    current = current.reshape((cycles, 2 * T_p))
    current = current.mean(axis=1)
    return current


x = np.arange(N_p) % (2 * N_x)
current = walk_saw_tooth_current(x.copy())
print("Cycle-averaged current for ratched potensial, alpha=0.8:")
print(current)
print()
print("Cycle-averaged current for ratched potensial, alpha=0.1:")
current = walk_saw_tooth_current(x.copy(), alpha=0.1)
print(current)
```

Grunner til at vi her får ikke-null strømming er fordi at partiklene vil bevege seg mot bunnpunktene når potensialet ikke er konstant. Når vi har et ikke-symmetrisk hakkete potensial, vil et større område av dette potensialet helle mot en spesifikk retning. I det konstante potensialet vil partiklene igjen spre seg og prossessen gjentas.

Vi ser at gjennomsnittsstrømmene for den første syklusen ($J_{avg}(0)$) har større magnitude enn de resterende strømmene. Dette er fordi at partiklene er spredt jevnt på starten og vil derfor "dette" ned hele potensialet til bunnen, mens etter hvert vil partiklene oppholde seg i bunnen og spre seg derifra når potensialet bytter til konstant og dermed ikke ha like gode forutsetninger for å lage høy strøm i en retning. Observerer i tillegg at alpha>0.5 gir gjennomsnittlig negativ strømming, og motsatt for alpha<0.5.

Når alpha nærmer seg enten 0 eller 1 vil strømmen bli større ettersom mindre og mindre del av potensialet peker i "feil" retning. Dette vises tydelig i at strømmen når alpha=0.1 er større enn ved alpha=0.8.

De trenger et visst antall iterasjoner for å komme seg over kneika. Vi ser også at man kommer til et punkt der å øke T_p ikke lenger øker strømmen, ettersom hvis partiklene er godt nok spredt på det konstante potensialet vil det å øke T_p kun gjøre at samme strøm blir delt på et større antall iterasjoner, dermed blir gjennomsnittet mindre. 

# Oppgave 3c)

```python
import scipy.special as sp


def iterate_alpha_saw_tooth(alpha_vals: np.ndarray, beta_k: float = 1000):
    currents: list[float] = []
    alpha = 0.8
    for alpha in alpha_vals:
        x = np.concat(
            [np.full((20 * N_x), 0), np.full((20 * N_x), N_x)]
        )
        current = walk_saw_tooth_current(x, alpha=alpha, cycles=1, beta_k=beta_k)
        currents.append(current[0])
    return currents


def analytical_solution(
    alpha: np.ndarray | float, T_p: np.ndarray | float = 500, N_x=100
):
    """Analytisk løsning av strømming."""
    return (
        N_x
        / (4 * T_p)
        * (
            sp.erfc(alpha * N_x / 2 * np.sqrt(3 / T_p))
            - sp.erfc((1 - alpha) * N_x / 2 * np.sqrt(3 / T_p))
        )
    )


alpha_vals = np.linspace(0, 1, 50, endpoint=False)

currents = iterate_alpha_saw_tooth(alpha_vals)


plt.figure(figsize=(10, 6))
plt.plot(alpha_vals, currents, "b-o", markersize=4, label="Numerical")

plt.plot(
    alpha_vals,
    analytical_solution(alpha_vals),
    "r-o",
    markersize=4,
    label="Analytical"
)
plt.xlabel("alpha")
plt.ylabel("Average Current")
plt.title("Average Current vs alpha")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

Vi ser at den analytiske samsvarer svært godt med den numeriske. Vi ser også at de største strømmene skjer når alpha nærmer seg 0 eller 1, slik som forutsett tidligere. Alpha rundt 0.5 vil ikke gi noe strømming, også som forventet. Alpha beskriver hvor stor del av potensialet som går nedover i en bestemt retning, i vårt tilfelle beskriver alpha hvor stor del av kurven som  går nedover mot venstre. Dermed vil alpha lik 0 bety at hele kurven til potensialet går nedover til høyre altså at partiklene går mot høyre. For alpha=1 vil det motsatte skje. Alpha lik 0.5 tilsvarer en potensialkurve der like mye går nedover mot venstre som nedover mot høyre, altså ingen gjennomsnittlig strømming.

# Oppgave 3f)

Den analytiske løsningen tar utgangspunkt i at partiklene alltid beveger seg nedover potensialet, noe som kun gjelder for $k_BT \ll |V(x+1)-V(x)|$. Som kan skrives som $\beta k\gg |V(x+1)-V(x)|$, altså akkurat det vi observerte i 3d. Den analytiske løsningen er derfor bedre for en større $\beta k$, eller mindre termisk energi, ettersom dette fører til mer deterministiske sannsynligheter for bevegelsen til partiklene, hvilket var en sentral antagelse i utledningen av den analytiske løsningen. Den delen av argumentasjonen som inneholdt en implisitt antagelse var da de sa at partiklene alltid vil gå mot det lokale minima, dette holder ikke hvis den termiske energien er høy nok.

```python
@njit
def _walk_collision_saw_tooth(
    x: np.ndarray,
    walk_left: np.ndarray,
    walk_right: np.ndarray,
    idxs: np.ndarray,
    b: float,
    N_x: int,
    size: int,
    wrap: bool = True,
):
    """Går gjennom hver partikkel og ser om den har lyst til å gå en retning og sjekker om den kan gå den retningen."""
    np.random.shuffle(idxs)
    field_size = N_x * size
    for idx in idxs:
        if walk_left[idx]:
            # sjekker kollisjon med alle partikler i systemet
            dx = x[idx] - x
            a = dx % field_size
            a = np.minimum(a, field_size - a)
            sign = np.sign(dx) * ((-1) ** (np.abs(dx) // (field_size / 2)))
            collision = np.any((a <= b + 1) & (sign > 0))
            if not collision:
                x[idx] -= 1

        if walk_right[idx]:
            # sjekker kollisjon med alle partikler i systemet
            dx = x[idx] - x
            a = dx % (N_x * size)
            a = np.minimum(a, N_x * size - a)
            sign = np.sign(dx) * ((-1) ** (np.abs(dx) // (field_size / 2)))
            collision = np.any((a > 0) & (a <= b + 1) & (sign < 0))
            if not collision:
                x[idx] += 1
        if wrap:
            # Trenger ikke wrap når vi regner ut strømmen.
            x = wrap_around(x, N_x=N_x, size=size)
    return x


@njit
def walk_collision_sawtooth(
    x: np.ndarray,
    T_p: int = 500,
    cycles: int = 10,
    alpha: float = 0.8,
    beta_k: float = 1000,
    N_x: int = 100,
    b: float = 1,
    size: int = 4,
    wrap: bool = True,
) -> np.ndarray:
    """Itererer alle tidssteg og oppdaterer partikkelposisjon underveis."""
    _walk_dist_random = np.random.uniform(0, 1, size=(2 * T_p * cycles, x.shape[0]))
    idxs = np.arange(x.shape[0])
    x_out = np.zeros((2 * T_p * cycles, x.shape[0]))
    for step, walk_dist_random in enumerate(_walk_dist_random):
        t = step * dt
        if (t // T_p) % 2 == 0:
            walk_left = walk_dist_random < 1 / 3
            walk_right = walk_dist_random > 2 / 3
        else:
            bV = potential_func_sawtooth(x, beta_k, alpha, N_x)
            bV_pos = potential_func_sawtooth(x + 1, beta_k, alpha, N_x)
            bV_neg = potential_func_sawtooth(x - 1, beta_k, alpha, N_x)
            walk_left, walk_right = get_walk_arrs(bV, bV_pos, bV_neg, walk_dist_random)

        x = _walk_collision_saw_tooth(
            x, walk_left, walk_right, idxs, b, N_x=N_x, size=size, wrap=wrap
        )
        # Tar vare på alle posisjoner.
        x_out[step] = x.copy()
    return x_out


def break_on_wrap_1d(x: np.ndarray):
    """
    Hjelpefunksjon som legger inn nan (not a number) for at plottet ikke skal legge inn linjer når partikkelen hopper rundt kanten.
    """
    threshold = 0.5 * (np.max(x) - np.min(x))

    dx = np.diff(x, axis=0)
    wrap = np.abs(dx) > threshold

    x_plot = x.copy()
    x_plot[1:][wrap] = np.nan
    return x_plot


x = np.arange(0, 20 * 4, 10)
x_full = walk_collision_sawtooth(
    x, T_p=40, cycles=5, alpha=0.2, N_x=20, b=2, size=4, beta_k=1000
)
plt.figure()
plt.hlines(y=np.arange(0, 400, 40), xmin=0, xmax=80, colors="#00000062")
plt.plot(break_on_wrap_1d(x_full), np.arange(0, x_full.shape[0]))
plt.xlabel("Position")
plt.ylabel("Time step")
plt.title("Time v Position")
plt.show()
```

I starten ser vi at partiklene virrevandrer ettersom de er i et konstant potensial. Etter 40 tidssteg slås det hakkete potensialet på og vi ser at partiklene detter ned til de lokale minima. Ettersom alpha=0.2 ser vi at de fleste partiklene går til høyre for å oppnå dette lokale minima. Ettersom $\beta k$ er lik 1000 har vi såpass lav termisk energi at systemet er tilnærmet deterministisk når det hakkete potensialet slås på. Derfor ser vi at det tar få steg før partiklene møtes i det lokale minima og ettersom de ikke kan befinne seg på samme posisjon så står alle partiklene helt stille frem til potensialet igjen blir konstant og virrevandringen starter. Denne prosessen gjentas flere ganger, og for hver gang potensialet er konstant så sprer partiklene seg symmetrisk ut. Men ettersom det hakkede potensialet ikke er symmetrisk er sannsynligheten større for å havne på den siden av potensialet som har en svak stigning ned mot høyre. Vi ser derfor at noen partikler ender opp i et minima som er mer til høyre enn det de startet i.

# Oppgave 4b)

```python
def num_particles(N_s: int, N_x: int, b: float, rho: float):
    return int(rho * N_s * N_x / b)


def calc_currents_sawtooth_collision(
    rho_vals: np.ndarray,
    T_p: int = 300,
    cycles: int = 100,
    alpha: float = 0.2,
    N_x: int = 100,
    b: float = 20,
    size: int = 10,
    beta_k: float = 1000,
):
    """Regner ut gjennomsnittsrømmingen for hakkete potensial med kollisjon."""
    currents: list[float] = []
    for rho in rho_vals:
        particles = num_particles(size, N_x, b, rho)
        if particles == 0:
            currents.append(0)
            continue
        x = np.linspace(0, N_x * size, particles)
        x_full = walk_collision_sawtooth(
            x,
            T_p=T_p,
            cycles=cycles,
            alpha=alpha,
            N_x=N_x,
            b=b,
            size=size,
            beta_k=beta_k,
            wrap=False,
        )
        x_full = x_full.reshape((cycles, 2 * T_p, particles))
        currents.append(np.diff(x_full, axis=1).mean())
    return currents


rho_vals = np.linspace(0, 1, 10)
currents = calc_currents_sawtooth_collision(rho_vals)

plt.figure()
plt.plot(rho_vals, currents)
plt.xlabel("Particle density (rho)")
plt.ylabel("Average current")
plt.title("Current v particle density")
plt.grid(True, alpha=0.3)
plt.show()

```

Vi ser at når tettheten er nerme 0 og 1 har vi liten gjennomsnittlig strømming ettersom at ved for høy tetthet kan ikke partiklene flytte seg mye siden de bare kolliderer med hverandre. Ved for lav tetthet vil partiklene ofte ende opp alene nederst i en brønn og vil dermed ha en lengre vei å gå for å komme til neste brønn. Vi ser at det optimale er en mellomting.
